```python
# Import library yang umum digunakan

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# mendefinisikan nama kolom
header = ['name', 'landmass', 'zone', 'area', 'population', 'language', 'religion', 'bars', 'stripes', 'colours', 'red', 'green', 'blue', 'gold', 'white',
          'black', 'orange', 'mainhue', 'circles', 'crosses', 'saltires', 'quarters', 'sunstars', 'crescent', 'triangle', 'icon', 'animate', 
          'text', 'topleft', 'botright']

# membaca data
data = pd.read_csv('flag.data', header=0, names=header)
print (data.info())
```

```python
# Mengconvert data kategorik
data = pd.get_dummies(data=data, columns=['mainhue', 'topleft', 'botright'])
```

```python
# melihat sebagian data
data.head(10)
```

```python
# melihat ukuran atau banyaknya data
data.shape
```

```python
print('Jumlah data null di dalam dataset setiap kolom:\n', data.isnull().sum())
print("-"*50)

data.describe()
```

```python
# grafik mengenai kolom language
sns.set_style('whitegrid')
sns.countplot(x='language',data=data,palette='rainbow')
```

```python
# menghilangkan kolom yang berisi string dan tidak penting
data.drop(["name", 'area', 'population'], axis=1, inplace=True)

# melakukan pembagian data test dan data training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop(['language'],axis=1), 
                                                    data['language'], test_size=0.10, 
                                                    random_state=40)
```

```python
# melakukan pemodelan
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
# melakukan prediksi terhadap data test
predictions = logmodel.predict(X_test)

X_test.head()
```

```python
# hasil dari prediksi
predictions
```

```python
# output seharusnya
y_test
```

```python
# nilai skor dari model training
logmodel.score(X_train, y_train)
```

```python
# nilai skor dari model test
logmodel.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
